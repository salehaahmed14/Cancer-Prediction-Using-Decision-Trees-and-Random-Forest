Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Load Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

Exploring Data

```python
cancer['feature_names']
```

```python
print(cancer['DESCR'])
```

```python
cancer['target']
```

Conversion into DataFrame

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.info()
```

```python
df.describe()
```

Adding the Target Value to the DataFrame

```python
df['Cancer']=pd.DataFrame(cancer['target'])
df.head()
```

Creating Boxplots

```python
l=list(df.columns[0:10])
for i in range(len(l)-1):
    sns.boxplot(x='Cancer',y=l[i], data=df, palette='winter')
    plt.figure()
```

**Training and Prediction**

Train Test Split

```python
feature_only_df = df.drop('Cancer',axis=1)
feature_only_df.head()
```

```python
target_df = df['Cancer']
target_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(feature_only_df,target_df,test_size = 0.30, random_state = 101)
y_train.head()
```

**Implementing Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion = 'gini',max_depth = None)
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
confusion_matrix = confusion_matrix(y_test,predictions)
print(confusion_matrix)
print ("Accuracy of prediction:",round((confusion_matrix[0,0]+confusion_matrix[1,1])/confusion_matrix.sum(),3))
```
